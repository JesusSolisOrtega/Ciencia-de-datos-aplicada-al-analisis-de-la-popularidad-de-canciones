```python

import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import json
import csv
import time
import re
import requests
import warnings
warnings.filterwarnings('ignore')

file_path = 'docs/600k_tracks.csv' 

# Configuración para visualizaciones en Jupyter Notebook
%matplotlib inline
sns.set(style="whitegrid")
pd.set_option('display.max_columns', None)

df = pd.read_csv(file_path)
df.head()
```

```python
initial_df = pd.read_csv('docs/600k_tracks.csv')

exclude_columns = ['id', 'name', 'id_artists', 'artists', 'release_date']

filtered_df = initial_df[[col for col in initial_df.columns if col not in exclude_columns]]

filtered_df.to_csv('docs/600k_tracks_anonimous.csv', index=False)

filtered_df.head()
```

```python
with open('docs/raw_responses.json', 'r') as file:
    raw_responses = file.readlines()

responses_dfs = []

for response in raw_responses:
    response_dict = json.loads(response)
    
    if response_dict:
        flattened_data = {
            'id': response_dict.get('id'),
            'popularity': response_dict.get('popularity'),
        }
        
        responses_dfs.append(flattened_data)

responses_df = pd.DataFrame(responses_dfs)

df_with_popularity = pd.merge(df, responses_df, how='left', left_on='id', right_on='id')

df_with_popularity.to_csv('docs/1-2M_with_popularity.csv', index=False)

df_with_popularity.head()
```

```python
batch_size = 1000

with open('docs/raw_responses.json', 'r') as file:
    raw_responses = file.readlines()

flattened_data_list = []

for i in range(0, len(raw_responses), batch_size):
    batch_responses = raw_responses[i:i + batch_size]

    for response in batch_responses:
        response_dict = json.loads(response)

        if response_dict:
            flattened_data = {
            'id': response_dict.get('id'),
            'album_type': response_dict.get('album', {}).get('album_type'),
            'artists': response_dict.get('album', {}).get('artists', [{}]),
            'available_markets_album': response_dict.get('album', {}).get('available_markets'),
            'external_urls_spotify_album': response_dict.get('album', {}).get('external_urls', {}).get('spotify'),
            'href_album': response_dict.get('album', {}).get('href'),
            'id_album': response_dict.get('album', {}).get('id'),
            'images_album': response_dict.get('album', {}).get('images'),
            'name_album': response_dict.get('album', {}).get('name'),
            'release_date': response_dict.get('album', {}).get('release_date'),
            'release_date_precision_album': response_dict.get('album', {}).get('release_date_precision'),
            'total_tracks_album': response_dict.get('album', {}).get('total_tracks'),
            'type_album': response_dict.get('album', {}).get('type'),
            'uri_album': response_dict.get('album', {}).get('uri'),
            'disc_number': response_dict.get('disc_number'),
            'duration_ms': response_dict.get('duration_ms'),
            'explicit': response_dict.get('explicit'),
            'external_ids_isrc': response_dict.get('external_ids', {}).get('isrc'),
            'external_urls_spotify_track': response_dict.get('external_urls', {}).get('spotify'),
            'href_track': response_dict.get('href'),
            'id_track': response_dict.get('id'),
            'is_local_track': response_dict.get('is_local'),
            'name_track': response_dict.get('name'),
            'popularity_track': response_dict.get('popularity'),
            'preview_url_track': response_dict.get('preview_url'),
            'track_number': response_dict.get('track_number'),
            'type_track': response_dict.get('type'),
            'uri_track': response_dict.get('uri'),
            'available_markets_track': response_dict.get('available_markets'),
            }

            flattened_data_list.append(flattened_data)

flattened_df = pd.DataFrame(flattened_data_list)

flattened_df = pd.merge(df, flattened_df, how='left', left_on='id', right_on='id')

flattened_df.to_csv('docs/flattened_dataset.csv', index=False)

flattened_df.head()
```
